```python
import pandas as pd
```

```python
yt = pd.read_csv('youtube.csv')
print(yt.head())
print()
print(yt.info())
print()
print(yt.describe())
```

```python
print(yt.isnull().sum())
yt = yt.drop_duplicates()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=yt, x='views', bins=30, kde=True)
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.pairplot(yt[['views', 'likes', 'dislike', 'comment_count']])
plt.show()

correlation_matrix = yt.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=yt, x='category_id')
plt.title('Video Category Distribution')
plt.xticks(rotation=90)
plt.xlabel('category ID')
plt.ylabel('count')
plt.show()
```
