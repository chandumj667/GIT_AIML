# 2: Write a python program to implement binary search to find the target values from the list:
    # a) Create a separate function to do binary search
    # b) Get the number of inputs from the user
    # c) Store the inputs individually in a list
    # d) In binary search function at first sort the list 
    # in order to start the search from the middle of the list
    # e) Compare the middle element to right and left elements to  search targwt element
    # f) If greater, move to right of list or else move to another side of the list
    # g) Print the result along with the position of the element

```python
def binary_search(nums, target, n):
    left =0
    right =n-1
#     found = False
    
    while left<=right:
        mid = (left+right)//2
        if nums[mid] == target:
            return mid+1
        elif nums[mid] > target:
            right = mid-1
        elif nums[mid] < target:
            left = mid+1
            
    raise Exception("Element not found in the list")
    
    
if __name__ == "__main__":
    n = int(input("Enter the size of the list: "))
    
    nums = []
    
    for i in range(n):
        ele = int(input("Enter element: "))
        nums.append(ele)
    
    nums.sort()
    print(nums)
    target = int(input("Enter the element to be searched: "))
    
    try:
        ind  = binary_search(nums, target, n)
        print(f"\nElement {target}  is found in the list at position {ind}")
    except Exception as e:
        print(f"Exception: {e}")
```

# 6: Write a python script to perform the following operations on a singly linked list
    a) Create a List
    b) Find the smallest element from the list
    c) Insert an element if it is not a duplicate element
    d) Display the elements in reverse order

```python
# class for node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
        
# class for creation of linkedlist
class Linkedlist:
    def __init__(self):
        self.head = None
    
#   inserting an elemeent into the list
    def append(self, data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
        else:
            cur = self.head
            while cur.next:
                cur = cur.next
            cur.next = new_node
                    
#   finding the smallest element  in the list
    def smaller_ele(self):
        if self.head == None:
            return None  
        
        mini = self.head.data
        cur = self.head.next

        while cur:
            if cur.data < mini:
                mini = cur.data
            cur = cur.next

        return mini

#   checking if the inserting element already present or not 
    def if_duplicate(self, data):
        new_node = Node(data)

        cur = self.head
        prev = None
        while cur:
            if cur.data == data:
                return  
            prev = cur
            cur = cur.next
        prev.next = new_node
        
        
        
#   reversing the linkedList
    def reverse(self):
        prev = None
        cur = self.head

        while cur:
            nxt_node = cur.next
            cur.next = prev
            prev = cur
            cur = nxt_node

        self.head = prev
        
#   display the nodes values
    def print(self):
        cur  = self.head
        while cur:
            print(f"{cur.data}-->", end="")
            cur = cur.next
        print("null")
    
    
# Creating the linkedlist by calling their respective methods
ll = Linkedlist()

while True:
    print("\nLinked List Operations:")
    print("1. Append")
    print("2. Find Smallest Element")
    print("3. Insert If Not Duplicate")
    print("4. Reverse")
    print("5. Display")
    print("6. Exit")

    choice = input("\nEnter your choice: ")

    if choice == '1':
        data = int(input("\nEnter data to append: "))
        ll.append(data)
    elif choice == '2':
        small = ll.smaller_ele()
        if small != None:
            print("\nSmallest element:", small)
        else:
            print("Linked list is empty. there are no elements in the list")
    elif choice == '3':
        val = int(input("Enter value to insert if not duplicate: "))
        ll.if_duplicate(val)
    elif choice == '4':
        ll.reverse()
        print("Linked List Reversed.")
        ll.print()
    elif choice == '5':
        ll.print()
    elif choice == '6':
        break
    else:
        print("Invalid choice. Please enter a valid option.")
```

# 7:Write a python program to implement the various operations for Stack ADT
    i) Push
    ii) Pop
    iii) Display

```python
class Stack:
    def __init__(self):
        self.elements = []

    def push(self, item):
        self.elements.append(item)

    def pop(self):
        if self.size() != 0:
            return self.elements.pop()
        else:
            raise IndexError("Pop from an empty stack")
    
    def size(self):
        return len(self.elements)  
    
    def display(self):
        if self.size() != 0:
            print("Stack elements:")
            for ele in reversed(self.elements):
                print(ele)
        else:
            print("Stack is empty")

if __name__ == "__main__":
    stk = Stack()

    while True:
        print("\nStack Operations:")
        print("1. Push")
        print("2. Pop")
        print("3. Display")
        print("4. Exit")

        choice = input("Enter your choice: ")
        if choice == '1':
            ele = input("Enter element to push: ")
            stk.push(ele)
            print(f"{ele} pushed onto the stack.")
        elif choice == '2':
            try:
                ele = stk.pop()  
                print(f"Popped element: {ele}")
            except IndexError as e:
                print(f"Error: {e}")
        elif choice == '3':
            stk.display()
        elif choice == '4':
            break
        else:
            print("Invalid choice. Please enter a valid option.")
```

# 8: Write a python script to implement the various operations for Queue ADT
    i) Insert
    ii) Delete
    iii) Display

```python
class Queue:
    def __init__(self):
        self.elements = []

    def insert(self, ele):
        self.elements.append(ele)

    def delete(self):
        if self.size() != 0:
            return self.elements.pop(0)
        else:
            raise IndexError("Delete from an empty queue")

    def size(self):
        return len(self.elements)

    def display(self):
        if self.size() != 0:
            print("Queue elements:")
            for ele in self.elements:
                print(ele)
        else:
            print("Queue is empty")



if __name__ == "__main__":
    queue = Queue()

    while True:
        print("\nQueue Operations:")
        print("1. Insert")
        print("2. Delete")
        print("3. Display")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            ele = input("Enter element to insert: ")
            queue.insert(ele)
            print(f"{ele} inserted into the queue.")
        elif choice == '2':
            try:
                ele = queue.delete()
                print(f"Deleted element: {ele}")
            except IndexError as e:
                print(f"Error: {e}")
        elif choice == '3':
            queue.display()
        elif choice == '4':
            break
        else:
            print("Invalid choice. Please enter a valid option.")
```

# 9: Write a python program to concert the following infix expressions to its postfix from using and pop operations of a stack
    a) A/B^C+D*E-F*G
    b) (B^2-4*A*C)^(1/2) (100)

```python
def infix_to_postfix(expression):
    def precedence(operator):
        precedence_dict = {"+": 1, "-": 1, "*": 2, "/": 2, "^": 3}
        return precedence_dict.get(operator, 0)

    def is_operator(char):
        return char in "+-*/^"

    def higher_precedence(op1, op2):
        return precedence(op1) > precedence(op2)

    postfix = []
    stack = []

    for char in expression:
        if char.isalnum():  
            postfix.append(char)
        elif char == "(":  
            stack.append(char)
        elif char == ")":  
            while stack and stack[-1] != "(":
                postfix.append(stack.pop())
            stack.pop()  
        elif is_operator(char):  
            while stack and stack[-1] != "(" and higher_precedence(stack[-1], char):
                postfix.append(stack.pop())
            stack.append(char)

    while stack:
        postfix.append(stack.pop())

    return "".join(postfix)


# Example usage:
infix_expr1 = "A/B^C+D*E-F*G"
infix_expr2 = "(B^2-4*A*C)^(1/2) (100)"

postfix_expr1 = infix_to_postfix(infix_expr1)
postfix_expr2 = infix_to_postfix(infix_expr2)

print("Infix Expression 1\t:\t", infix_expr1)
print("Postfix Expression 1\t:\t", postfix_expr1)

print("\nInfix Expression 2\t:\t", infix_expr2)
print("Postfix Expression 2\t:\t", postfix_expr2)
```

# 3:Write a python program for sorting a list of elements using selection sort algorithm
    a) Assume two lists: Sorted list - intially empty and unsorted list- given input list
    b) In the first iteration, find the smallest element in the unsorted list and place it in the sorted list
    c) In the second iteration, find the smallest element in the unsorted list 
    and place it in the correct position by compariong with the element in the sorted lisst
    d) In the third iteration, again find the smallest element in the unsorted list 
    and place it in the correct position by comparing till the unsorted list becomes empty
    e) This process continues till the unsorted list becomes empty
    f) Display the sorted list

```python
def selection_sort(nums, n):

    for i in range(n - 1):
        ind = i

        for j in range(i + 1, n):
            if nums[j] < nums[ind]:
                ind = j

        nums[i], nums[ind] = nums[ind], nums[i]

nums= []
n = int(input("Enter the size of the list: "))
for i in range(n):
    ele =int(input("Enter the element: "))
    nums.append(ele)
print(f"\nUnsorted array is\t: {nums}")
selection_sort(nums, n)
print(f"\nSorted array is\t\t: {nums}")
```

# 4: Write a pytho program for sorting a list of elements using insertion sort algorithm:
    a) Assume two lists: Sorted list - Initially empty and Unsorted list - Given input list 
    b) In the first Iteration, take the first element in the unsorted list and insert it in sorted list
    c) In the second iteration, take the second element in the given list 
    and compare with the element in the sorted sub list and place it in the correct position
    d) In the third iteration, take the third element in the given list 
    and compare with the elements in the sorted sub list and place the elements in the correct position
    e) This process continures untill the last element is inserted in the sorted sublist
    f) Display the sorted elements


```python
def insertion_sort(nums):
    for i in range(1, len(nums)):
        value = nums[i]  
        j = i - 1 
        
        while ((value < nums[j]) and (j>=0)):
            nums[j + 1] = nums[j]
            j -= 1

        nums[j + 1] = value

        
n = int(input("Enter the size: "))
nums = []
for i in range(n):
    ele = int(input("Enter the element: "))
    nums.append(ele)

print(f"Unsorted List is:\t{nums}")
insertion_sort(nums)
print(f"Sorted list is:\t\t{nums}")
```

# 5: Write a python program that performs merge sort on a list of numbers:
    a) Divide : If the given array has zero or one element, return.
        i) Otherwise
        ii)Divide the input list in to two halves each containing half os the elements i.e. left half and right half
    b) Conquer: Recursively sort the two sub-arrays (left half and right half)
        i) Call the merge sort on left half
        ii)Call the merge sort on right half
    c) Combine: Combine the elements back in the input list by merging the two sorted lists into a sorted sequnce.

```python
def merge_sort(nums):
    if len(nums) > 1:
        mid = len(nums) // 2
        
#       a)Divide the array into two halves
        
        left = nums[:mid]  
        right = nums[mid:]

#       b)Recursive calls to merge_sort for both halves
        merge_sort(left)
        merge_sort(right)

        i = j = k = 0

#       c)Merge the two halves back into the original array
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                nums[k] = left[i]
                i += 1
            else:
                nums[k] = right[j]
                j += 1
            k += 1

#       remaining elemnts if in left or right then they are added in to the result
        while i < len(left):
            nums[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            nums[k] = right[j]
            j += 1
            k += 1


if __name__ == "__main__":
    try:
        nums = input("Enter the elements in to the list:\n").split()
        nums = [int(num) for num in nums]
    except Exception as e:
        print(f"Value Errot: {e}")
        
    print(f"Original array is:\t {nums}")
    merge_sort(nums)
    print(f"Sorted array is:\t {nums}")
```
