# 1: Make a bar plit of the months in which movies with "Christmas" in their title tend to be released in the USA

```python
christmas_movies_usa = release_dates[(release_dates['title'].str.contains('Christmas', case=False)) & (release_dates['country'] == 'USA')].copy()

christmas_movies_usa.reset_index(inplace=True)

christmas_movies_usa['month'] = christmas_movies_usa['date'].dt.month

month_counts = christmas_movies_usa['month'].value_counts().sort_index()

all_months = range(1, 13)
month_counts = month_counts.reindex(all_months, fill_value=0)

month_names = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

plt.figure(figsize=(10, 6))
plt.bar(month_names, month_counts, width=0.4)
plt.xlabel('Month')
plt.ylabel('Number of "Christmas" Movies')
plt.title('Number of "Christmas" Movies Released in Each Month in the USA')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 6: In which months do films with Judi Dench tend to be released in the USA?

```python
judi_dench_films = cast[cast['name'] == 'Judi Dench']
judi_dench_films = pd.merge(judi_dench_films, release_dates, on=['title'])

judi_dench_films['date'] = pd.to_datetime(judi_dench_films['date'])

judi_dench_films['month'] = judi_dench_films['date'].dt.month

month_counts = judi_dench_films['month'].value_counts().sort_index()
month_names = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

plt.figure(figsize=(10, 6))
plt.bar(month_names, month_counts, width=0.4)
plt.xlabel('Month')
plt.ylabel('Number of Films featuring Judi Dench')
plt.title('Number of Films featuring Judi Dench Released in Each Month in the USA')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```
