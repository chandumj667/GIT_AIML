# 1: What are the ten most common movies names of all time?

```python
common_mvs = titles["title"].value_counts().head(10)

print(f"The ten most common movie titles are:\n\n{common_mvs}")
```

# 2: Which three years of the  1930s saw the most films released?

```python
in_1930s = titles[(titles["year"]>=1930) & (titles["year"]<=1939)]

per_year = in_1930s["year"].value_counts()

three = per_year.head(3)

print(f"The threee years of 1930s with most films:\n{three}")
```

# 3: Plot the number of films that have been released each decade over the history of cinema

```python
import matplotlib.pyplot as plt
films_decade = titles.groupby((titles["year"]//10)*10).size()

films_decade.plot(kind = "bar", figsize=(12, 6))

plt.xlabel("decade")
plt.ylabel("number of films released")
plt.title("Number of films released Each Decade")
plt.xticks(rotation = 45)
plt.show()
```

# 15: Plot the n-values of the roles that Judi Dench has played over her career

```python
judi = cast[(cast["name"] == "Judi Dench") & (~cast["n"].isna())]

plt.figure(figsize=(12, 6))
plt.scatter(judi['year'], judi['n'], alpha=0.5)
plt.xlabel('Year')
plt.ylabel('n-Value')
plt.title('n-Values of Roles Played by Judi Dench Over Her Career')
plt.grid(True)
plt.show()
```

# 18: How many leading roles were available to actors andd how many actress in the 1950s

```python
role_1950s = cast[(cast['year'] >= 1950) & (cast['year'] <= 1959)]

leading_actors = len(role_1950s[(role_1950s['type'] == 'actor') & (role_1950s['n'] == 1)])
leading_actresses = len(role_1950s[(role_1950s['type'] == 'actress') & (role_1950s['n'] == 1)])

print(f"Number of leading roles for actors in the 1950s: {leading_actors}")
print(f"Number of leading roles for actresses in the 1950s: {leading_actresses}")
```

# 19: How many supporting roles were avaialable to actors and how many to actress in th e1950s?

```python
role_1950s = cast[(cast['year'] >= 1950) & (cast['year'] <= 1959)]

sup_actors = len(role_1950s[(role_1950s['type'] == 'actor') & (role_1950s['n'] == 2)])
sup_actresses = len(role_1950s[(role_1950s['type'] == 'actress') & (role_1950s['n'] == 2)])

print(f"Number of supporting roles for actors in the 1950s: {sup_actors}")
print(f"Number of supporting roles for actresses in the 1950s: {sup_actresses}")
```
