# 1:Define a year as a "Superman Year" Whose films feature more superman Characters than Batman. How many years in film history have been superman years?

```python
superman = cast['character'].str.contains('Superman', case=False, na=False)

batman = cast['character'].str.contains('Batman', case=False, na=False)

sm_bm = cast[superman | batman].groupby(['year', 'title']).agg({'character': 'count'})

sm_batman_sum = sm_bm.groupby('year').agg({'character': 'sum'})

sm_years = sm_batman_sum[sm_batman_sum['character'] > 0]

num_sm_years = len(sm_years)
print(f"Number of Superman Years: {num_sm_years}")
```

# 2:Plot the number of actor roles each year and the number of actress roles each year over the history of film

```python
roles_by_year = cast.groupby(['year', 'type']).size().unstack()
roles_by_year.plot(kind='line', figsize=(12, 6))
plt.xlabel('Year')
plt.ylabel('Number of Roles')
plt.title('Number of Actor and Actress Roles Each Year')
plt.show()
```

# 3:Plot the number of actor roles each year and the number of acress roles each year, but this time as a kind ="area" plot.

```python
roles_by_year.plot(kind='area', figsize=(12, 6))
plt.xlabel('Year')
plt.ylabel('Number of Roles')
plt.title('Number of Actor and Actress Roles Each Year (Area Plot)')
plt.show()
```

# 4:plot the difference between the number of actor roles each year and then number of actress roles each year over the history of film

```python
roles_by_year.plot(kind='area', figsize=(12, 6))
plt.xlabel('Year')
plt.ylabel('Number of Roles')
plt.title('Number of Actor and Actress Roles Each Year (Area Plot)')
plt.show()
```

# 5:plot the difference between the number of actor roles each yeaar and the  number of actress roles each year over the history of film

```python
role_difference = roles_by_year['actor'] - roles_by_year['actress']

role_difference.plot(kind='line', figsize=(12, 6))
plt.xlabel('Year')
plt.ylabel('Difference (Actor - Actress)')
plt.title('Difference Between Actor and Actress Roles Each Year')
plt.show()
```

# 6:plot the fraction of roles that have been "actor" roles each year in the history of film

```python
fraction_actor_roles = roles_by_year['actor'] / (roles_by_year['actor'] + roles_by_year['actress'])

fraction_actor_roles.plot(kind='line', figsize=(12, 6))
plt.xlabel('Year')
plt.ylabel('Fraction of Actor Roles')
plt.title('Fraction of Actor Roles Each Year')
plt.show()
```

# 7: plot the fraction of supporting (n=2) roles that have been actior roles each year in the history of film

```python
supporting_roles_mask = (cast['n'] == 2)

supporting_roles_by_year = cast[supporting_roles_mask].groupby(['year', 'type']).size().unstack()

fraction_supporting_actor_roles = supporting_roles_by_year['actor'] / (supporting_roles_by_year['actor'] + supporting_roles_by_year['actress'])

fraction_supporting_actor_roles.plot(kind='line', figsize=(12, 6))
plt.xlabel('Year')
plt.ylabel('Fraction of Supporting Actor Roles (n=2)')
plt.title('Fraction of Supporting Actor Roles (n=2) Each Year')
plt.show()
```

# 8: Build a plot with a line for each rank n =1 through n=3, where the linke shows what fraction of that rank's roles were actor roles for each year in the history of film

```python
filtered_cast = cast[cast['n'].between(1, 3)]
pivot_table = filtered_cast.pivot_table(index=['year', 'n'], columns='type', aggfunc='size', fill_value=0)

for n in range(1, 4):
    actor_roles = pivot_table.loc[(slice(None), n), 'actor']
    actress_roles = pivot_table.loc[(slice(None), n), 'actress']
    pivot_table[f'Fraction_actor_roles_n={n}'] = actor_roles / (actor_roles + actress_roles)

for n in range(1, 4):
    column_name = f'Fraction_actor_roles_n={n}'
    pivot_table[column_name].unstack().plot(kind='line', label=f'n={n}', figsize=(12, 6))

plt.xlabel('Year')
plt.ylabel('Fraction of Actor Roles')
plt.title('Fraction of Actor Roles by Rank (n) Each Year')
plt.legend(title='Rank (n)')
plt.show()
```
