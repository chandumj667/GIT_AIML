# 1: Write a python program to create a base class "Shape"

```python
# with methods to calculate area and perimeter.
# Then Create derived classes "Circle" and "Rectangle" 
# that inherit  from the base class and calculate their 
# respective area and perimeters.
# Demonstrate their usage in a program.
```

```python
import math

# Shape class
class Shape:
    def area(self):
        print("Area: ")

    def perimeter(self):
        print("Perimeter: ")
        

# Circle class which is derived from the Shape class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

    
# Rectangle class which is derived from the Shape class
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Demonstrate the usage of the classes
if __name__ == "__main__":
    
    try:

        print("For Circle\n")
        radius = int(input("Enter radius"))

        # Create a Circle object
        circle = Circle(radius)
        print("Circle - Radius:", circle.radius)
        print("Circle - Area:", circle.area())
        print("Circle - Perimeter:", circle.perimeter())

        print("\nFor Rectangle\n")
        length = int(input("Enter the length of the rectangle: "))
        width = int(input("Enter the width of the rectangle: "))

        # Create a Rectangle object
        rectangle = Rectangle(length, width)
        print("\nRectangle - Length:", rectangle.length)
        print("Rectangle - Width:", rectangle.width)
        print("Rectangle - Area:", rectangle.area())
        print("Rectangle - Perimeter:", rectangle.perimeter())
        
    except BaseException as e:
        print(f"Exception: {e}")
```

```python
# The system should be able to handle different types of accounts, 
# such as savings accounts and checking accounts.

# Both account types should have common attributes like an account number, account holders name and balance.
# However, savings accounts should have an additional attribute for interest rate 
# while checking accounts should have attribute for overdraft limit

# i) Create a python class called bank-account with the following attributes and methods:
#     a) Attributes: Account_number, account_holder_name and balance
#     b) Methods: __init__()(constructor), deposit(), and withdraw()
# ii) Create two subclasses, SavingsAccount and CheckingAccount, that inherit from the BankAccount class
# iii) Add the following attributes and methods to each subclass"
#     a) SavingsAccount:
#         i) Additional attribute: Interest_rate
#         ii)Method: Calculate_interest(), which calculate and add interest to the account base in the interest rate
#     b)CheckingAccount:
#         i)Additional attribute: overdraft_limit
#         ii)Method: withdraw allows withdrawung money up to the overdraft limit (if available) without additional fees
# iv)write a program that creates instances of both SavingsAccount and CheckingAccount 
# and demonstrates the use of their methods.
# v) Implement proper encapsulation by making the attributes private where necessary 
# and providing getter and setter methods as needed
# vi) Handle any potential errors or exceptions 
# that may occur during operations like withdrawls, deposits or interest calculations
# vii) Provide comments in your code to explain the purpose of each class, attribute and mehod

# Note: You code should create instances of the classes, simulate transactions 
# and showcase the differences between Savings  Accounts and Checking Accounts.
```

```python
# class for BankAccount and 
# it was inherited by the Savings and Checking Account classes
class BankAccount:
    
    def __init__(self, acc_num, name):
        self.__acc_num = acc_num  
        self.__name = name        
        self.__bal = 0.0          

#   getter method for account number
    def get_account_number(self):
        return self.__acc_num

#   getter method for person name
    def get_name(self):
        return self.__name

#   getter method for to check balance  
    def get_balance(self):
        return self.__bal
    
#   methof for deposit money in to the account
    def deposit(self, depo_amount):
        if depo_amount > 0:
            self.__bal += depo_amount
            print(f"\nDeposited amount in {self.get_account_number()} is {depo_amount} and the balance is: {self.__bal}\n")
        else:
            raise ValueError("Enter an amount greater than zero\n")
            
#   method for withdraw money from the account
    def withdraw(self, wd_amount):
        if 0 < wd_amount <= self.__bal:
            self.__bal -= wd_amount
            print(f"Withdrawn amount from {self.get_account_number()} is {wd_amount} and the balance is: {self.__bal}\n")
        else:
            raise ValueError("Insufficient balance or invalid withdrawl amount\n")

            
#Class for Saving Account and it is derived from class Bank Account
class SavingsAccount(BankAccount):
    def __init__(self, acc_num, name, interest_rate):
        super().__init__(acc_num, name)
        self.__interest_rate = interest_rate

#   interest calculation and deposit in to the account
    def calculate_interest(self):
        interest_yearly = self.get_balance() * 1 * self.__interest_rate
        print(f"Interest on savings account{super().get_account_number()}: {interest_yearly}\n")
        self.deposit(interest_yearly)


        
# Class for Checking Account and it is derived from Class Bank Account
class CheckingAccount(BankAccount):
    def __init__(self, acc_num, name, od_limit):
        super().__init__(acc_num, name)
        self.__od_limit = od_limit

    def withdraw(self, wd_amount):
        if wd_amount > 0 and (wd_amount <= (self.get_balance() + self.__od_limit)):
            super().withdraw(wd_amount)
            print(f"Withdrawn amount from {super().get_account_number()} is {wd_amount} and the remaining balance is: {self.get_balance()}\n")
        else:
            raise ValueError("Entered amount exceeds the overdraft limit or is invalid\n")

if __name__ == "__main__":
    p1name = input("Enter customer1 name: ")
    p1acc_num = input("Enter customer1 account number: ")
    psint = float(input("Enter interest rate for savings account: "))/100
    p1bal = float(input("Enter Balance1: "))
    
    p2name = input("Enter customer2 name: ")
    p2acc_num = input("Enter customer2 account number: ")
    pCodlimit = float(input("Enter overdraft limit for checking account: "))
    p2bal = float(input("Enter Balance2: "))
    
    sav_acc = SavingsAccount(p1acc_num, p1name, psint)
    sav_acc.deposit(p1bal)
    
    check_acc = CheckingAccount(p2acc_num, p2name, pCodlimit)
    check_acc.deposit(p2bal)
    
    sav_acc.calculate_interest()
    
    depo_sav = float(input("Enter deposit amount into savings account: "))
    sav_acc.deposit(depo_sav)
    
    wd_sav = float(input("Enter withdrawal amount from savings account: "))
    sav_acc.withdraw(wd_sav)
    
    depo_check = float(input("Enter deposit amount into checking account: "))
    check_acc.deposit(depo_check)

    wd_check = float(input("Enter withdrawal amount from checking account: "))
    check_acc.withdraw(wd_check)
    
    print("\nDetails of Customer1:")
    print("Name:", sav_acc.get_name())
    print("Account Number:", sav_acc.get_account_number())
    print("Balance:", sav_acc.get_balance())
    
    print("\nDetails of Customer2:")
    print("Name:", check_acc.get_name())
    print("Account Number:", check_acc.get_account_number())
    print("Balance:", check_acc.get_balance())
```

```python
# Class Employee and it was inherited by other two classes
class Employee:
    def __init__(self, name, emp_id, salary):
        self.__name = name
        self.__emp_id = emp_id
        self.__salary = salary

    
#   setters methods for attributes
    def set_name(self, name):
        self.__name = name

    def set_emp_id(self, emp_id):
        self.__emp_id = emp_id

    def set_salary(self, salary):
        self.__salary = salary
        
#   getters methids for attributes
    def get_name(self):
        return self.__name

    def get_emp_id(self):
        return self.__emp_id

    def get_salary(self):
        return self.__salary

    
#   method to return the salary
    def calculate_salary(self):
        return self.__salary


# Fulltimeemployee class derived form Employee class    
class FullTimeEmployee(Employee):
    def __init__(self, name, emp_id, annual_salary):
        super().__init__(name, emp_id, annual_salary)

    def calculate_salary(self):
        return self.get_salary() / 12
    

# parttimeemployee class derived form Employee class    
class PartTimeEmployee(Employee):
    def __init__(self, name, emp_id, no_of_hrs, per_hour):
        super().__init__(name, emp_id, None)
        self.__no_of_hrs = no_of_hrs
        self.__per_hour = per_hour

    def get_hours_worked(self):
        return self.__no_of_hrs

    def set_hours_worked(self, no_of_hrs):
        self.__no_of_hrs = no_of_hrs

    def get_hourly_rate(self):
        return self.__per_hour

    def set_hourly_rate(self, per_hour):
        self.__per_hour = per_hour

    def calculate_salary(self):
        return self.__no_of_hrs * self.__per_hour


if __name__ == "__main__":
    try:
        
        while True:
            print("\nchoose employee type\n1 : Full-Time\n2 : Part-Time\n0 : Exit")
            emp_type = input("Enter employee type : ")

            if emp_type == "0":
                print("Closing...")
                break
            elif (emp_type == "1" or emp_type == "2"): 
                
                name = input("Enter employee name: ")
                emp_id = input("Enter employee ID: ")


                if emp_type == "1":
                    annual_salary = float(input("Enter annual salary: "))
                    employee = FullTimeEmployee(name, emp_id, annual_salary)
                elif emp_type == "2":
                    no_of_hrs = float(input("Enter hours worked: "))
                    per_hour = float(input("Enter hourly rate: "))
                    employee = PartTimeEmployee(name, emp_id, no_of_hrs, per_hour)




                salary = employee.calculate_salary()
                print(f"Employee Name: {employee.get_name()}")
                print(f"Employee ID: {employee.get_emp_id()}")
                print(f"Calculated Salary: RS. {salary:.2f}")
            
            else:
                print("Invalid employee type.")

    except ValueError as e:
        print(f"Error: {e}")

```

```python
class Product:
    def __init__(self, prod_id, name, price, stock):
        self.__prod_id = prod_id
        self.__name = name
        self.__price = price
        self.__stock = stock

    def get_prod_id(self):
        return self.__prod_id

    def get_name(self):
        return self.__name

    def get_price(self):
        return self.__price

    def get_stock(self):
        return self.__stock

    def set_stock(self, stock):
        self.__stock = stock

    def is_available(self, quantity):
        return self.__stock >= quantity

    def calculate_price(self, quantity):
        if self.is_available(quantity):
            return self.__price * quantity
        else:
            raise Exception("Product is out of stock")

    def __str__(self):
        return f"{self.__name} - Rs.{self.__price:.2f}"


class Electronics(Product):
    def __init__(self, prod_id, name, price, stock, warranty_months):
        super().__init__(prod_id, name, price, stock)
        self.__warranty_months = warranty_months

    def get_warranty(self):
        return self.__warranty

    def __str__(self):
        return f"\nElectronics: {super().__str__()}, Warranty: {self.__warranty_months} months"


class Clothing(Product):
    def __init__(self, prod_id, name, price, stock, size):
        super().__init__(prod_id, name, price, stock)
        self.__size = size

    def get_size(self):
        return self.__size

    def __str__(self):
        return f"\nClothing: {super().__str__()}, Size: {self.__size}"


class FoodAndBeverages(Product):
    def __init__(self, prod_id, name, price, stock, expiration_date):
        super().__init__(prod_id, name, price, stock)
        self.__expiration_date = expiration_date

    def get_expiration_date(self):
        return self.__expiration_date

    def __str__(self):
        return f"\nFood and Beverages: {super().__str__()}, Expiration Date: {self.__expiration_date}"


class Healthcare(Product):
    def __init__(self, prod_id, name, price, stock, active_ingredient):
        super().__init__(prod_id, name, price, stock)
        self.__active_ingredient = active_ingredient

    def get_active_ingredient(self):
        return self.__active_ingredient

    def __str__(self):
        return f"\nHealthcare: {super().__str__()}, Active Ingredient: {self.__active_ingredient}"


class ShoppingCart:
    def __init__(self):
        self.__items = []

    def add_item(self, product, quantity):
        try:
            total_price = product.calculate_price(quantity)
            self.__items.append((product, quantity, total_price))
            product.set_stock(product.get_stock() - quantity)
            print(f"Added {quantity} {product.get_name()} to the cart.")
        except Exception as e:
            print(f"Error: {e}")

    def calculate_total_cost(self):
        total_cost = 0
        for item in self.__items:
            total_cost += item[2]
        return total_cost

    def display_cart(self):
        print("Shopping Cart:")
        for item in self.__items:
            product, quantity, total_price = item
            print(f"\n{product} x{quantity} = Rs.{total_price:.2f}")
#         print(f"Total Cost: Rs.{self.calculate_total_cost():.2f}")




if __name__ == "__main__":
    # Create a shopping cart
    cart = ShoppingCart()

    while True:
        print("\n Products:")
        print("1. Laptop Bags - Rs.1000.00")
        print("2. Polo T-Shirt - Rs.250.00")
        print("3. Soft Drink - Rs.75.00")
        print("4. Dolo 650 - Rs.20.00")
        print("5. Checkout and Exit")

        choice = input("Enter the product number to add to the cart (1-5): ")

        if choice == "1":
            quantity = int(input("Enter the quantity: "))
            cart.add_item(Electronics(1, "Laptop Bags", 1000.00, 5, 12), quantity)
        elif choice == "2":
            quantity = int(input("Enter the quantity: "))
            cart.add_item(Clothing(2, "Polo T-Shirt", 250.00, 10, "Large"), quantity)
        elif choice == "3":
            quantity = int(input("Enter the quantity: "))
            cart.add_item(FoodAndBeverages(3, "Soft drink", 75.00, 20, "2023-10-10"), quantity)
        elif choice == "4":
            quantity = int(input("Enter the quantity: "))
            cart.add_item(Healthcare(4, "Dolo 650", 20.00, 30, "Paracetmol"), quantity)
        elif choice == "5":
            cart.display_cart()
            print(f"\nTotal Cost: Rs.{cart.calculate_total_cost():.2f}")
            break
        else:
            print("Invalid choice. Please enter a valid product number (1-5).")


```

```python
# Scnario: you are developing a library management system 
# where you need to handle books, partrons, and Library transactions.

# 1: Create a class hierarchy that includes classes for books (e.g., Book), 
# patrons (e.g., Patron), and transactions (e.g., Transaction).
# Define attributes and methods for each class

# 2: Implement encapsulation by making relevant attributes private 
# and providing getter and setter methods where necessary

# 3: Use inheritance to represent different types of books (e.g., fiction, non-fiction) 
# as subclasses of the book class. Ensure that each book type can have specific attributes and methods

# 4:Demonstrate polymorphism by allowing patrons to check out and return books, regardless of the book type.

# 5: Implement a method for tracking overdue books and notifying patrons

# 6: Consider scenarios like book reservations, late fees and library staff interactions in your design
```

```python
import datetime

# Base class for books
class Book:
    def __init__(self, title, author):
        self.__title = title
        self.__author = author
        self.__checked_out = False

    def get_title(self):
        return self.__title

    def get_author(self):
        return self.__author

    def is_checked_out(self):
        return self.__checked_out

    def check_out(self):
        if not self.__checked_out:
            self.__checked_out = True
            due_date = datetime.date.today() + datetime.timedelta(days=14)
            return due_date
        else:
            raise ValueError("The book is already checked out")

    def return_book(self):
        if self.__checked_out:
            self.__checked_out = False
        else:
            raise ValueError("The book is not checked out")

# FictionBook subclass
class FictionBook(Book):
    def __init__(self, title, author, genre):
        super().__init__(title, author)
        self.__genre = genre

    def get_genre(self):
        return self.__genre

# NonFictionBook subclass
class NonFictionBook(Book):
    def __init__(self, title, author, topic):
        super().__init__(title, author)
        self.__topic = topic

    def get_topic(self):
        return self.__topic

# Transaction class
class Transaction:
    def __init__(self, patron, book, due_date):
        self.__patron = patron
        self.__book = book
        self.__due_date = due_date
        self.__returned = False

    def get_patron(self):
        return self.__patron

    def get_book(self):
        return self.__book

    def get_due_date(self):
        return self.__due_date

    def is_returned(self):
        return self.__returned

    def return_book(self):
        if not self.__returned:
            self.__returned = True
        else:
            raise ValueError("The book is already marked as returned")

# Patron class with transactions
class Patron:
    def __init__(self, name, patron_id):
        self.__name = name
        self.__patron_id = patron_id
        self.__books_checked_out = []
        self.__transactions = []

    def get_name(self):
        return self.__name

    def get_patron_id(self):
        return self.__patron_id

    def get_books_checked_out(self):
        return self.__books_checked_out

    def check_out_book(self, book):
        if book.is_checked_out():
            raise ValueError("The book is already checked out")
        else:
            due_date = book.check_out()
            self.__books_checked_out.append((book, due_date))

            # Create a transaction for the checked-out book
            transaction = Transaction(self, book, due_date)
            self.__transactions.append(transaction)

    def return_book(self, book):
        found = False
        for checked_book, _ in self.__books_checked_out:
            if checked_book == book:
                book.return_book()
                self.__books_checked_out.remove((book, _))
                found = True
                break

        if not found:
            raise ValueError("You did not check out this book")

if __name__ == "__main__":
    # Create a list of available books
    available_books = [
        FictionBook("The God of Small Things", "Arundhati Roy", "Contemporary Fiction"),
        FictionBook("A Suitable Boy", "Vikram Seth", "Historical Fiction"),
        FictionBook("The White Tiger", "Aravind Adiga", "Satire"),
        FictionBook("Midnight's Children", "Salman Rushdie", "Magical Realism"),
        FictionBook("The Namesake", "Jhumpa Lahiri", "Drama"),

        NonFictionBook("The Argumentative Indian", "Amartya Sen", "Politics"),
        NonFictionBook("India After Gandhi", "Ramachandra Guha", "History"),
        NonFictionBook("Sapiens: A Brief History of Humankind", "Yuval Noah Harari", "Anthropology"),
        NonFictionBook("Wings of Fire", "A.P.J. Abdul Kalam", "Autobiography"),
        NonFictionBook("The Discovery of India", "Jawaharlal Nehru", "Historical Text"),
    ]

    # Take user's name and ID
    user_name = input("Enter your name: ")
    user_id = input("Enter your ID: ")

    # Create a patron with the provided name and ID
    patron = Patron(user_name, user_id)

    # Display the list of available books
    print("Available books:")
    for i, book in enumerate(available_books, 1):
        print(f"{i}. {book.get_title()} by {book.get_author()} ({book.get_genre() if isinstance(book, FictionBook) else book.get_topic()})")

    while True:
        try:
            book_choice = int(input("\nEnter the number of the book you want to check out (0 to finish): "))
            if book_choice == 0:
                break
            elif 1 <= book_choice <= len(available_books):
                selected_book = available_books[book_choice - 1]
                if selected_book.is_checked_out():
                    print(f"The book '{selected_book.get_title()}' is already checked out.")
                else:
                    patron.check_out_book(selected_book)
                    print(f"\nYou have successfully checked out {selected_book.get_title()}!")
            else:
                print("Invalid book number.")
        except ValueError:
            print("\n***Invalid input. Please enter a valid book number.")
        
    # Display the books checked out by the patron
    print(f"\nBooks checked out by {patron.get_name()} (ID: {patron.get_patron_id()}):")
    for book, due_date in patron.get_books_checked_out():
        print(f"\n- {book.get_title()} by {book.get_author()}, Due Date: {due_date}")
```
